```python
kick = pd.read_csv('kickPoints.csv', header = None)
```

```python
kick.head()
```

```python
kick = kick.drop([35],axis = 1)
```

```python
data2 = kick.values[:,:]
```

```python
data = np.append(data,data2,axis = 0)
```

```python
#Data = []
```

```python
num_kick = len(data) - num_punch
num_kick
```

```python
labels = []
for i in range(len(data)):
    if(i<num_punch):
        labels.append(0)
    else:
        labels.append(1)
labels = np.asarray(labels)
```

```python
data_size = 35 # 1*35
num_classes = 2 # punch and kick
```

```python
labels.shape
```

```python
data = np.nan_to_num(data)
data = np.where(data == -1, 0, data)
```

```python
data[0:5,:]
```

```python
#train_data = []
#for i in range(len(Data)):
#    train_data.append(Data[i].flatten())
#train_data = np.asarray(train_data)
```

```python
labels = keras.utils.to_categorical(labels, num_classes)
labels
```

```python
model = Sequential()
```

```python
model.add(Dense(units=32, activation='relu', input_shape=(data_size,)))
model.add(Dense(units=num_classes, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(data, labels, epochs=200, batch_size=32,verbose=True)
```

```python
loss, accuracy  = model.evaluate(data, labels, verbose=True)
```

```python
print(loss,accuracy)
```
